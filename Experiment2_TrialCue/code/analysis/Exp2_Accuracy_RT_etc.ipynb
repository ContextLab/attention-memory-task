```python
import random

total = []
for sub in range(0,10000):
    
    corr = 0
    probe = ['x']*40 + ['o']*40
    random.shuffle(probe)
    
    for trial in probe:
        response = random.choice(['x','o'])
        if trial==response:
            corr +=1
    total.append(corr)
```

```python
plt.hist([i/80 for i in total])
```
