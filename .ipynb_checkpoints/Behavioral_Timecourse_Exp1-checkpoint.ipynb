# Obtain Raw Difference Scores

```python
#sub_split = combo.groupby(['Subject','Attention Level'],as_index=False)['Familiarity Rating'].mean()
sub_split['Novs'] = 0.0
sub_split['CuedCategory'] = sub_split['Cued Category']

for x in sub_split['Subject'].unique():
    # for each subject

    k = sub_split[sub_split['Subject']==x]
    novels = k[k['Attention Level']=='Novel']
    # select just that subject's data (all, and novel only)
    
    fp = ['Face','Place']
    for pair in list(itertools.combinations_with_replacement(fp, r=2)):
    # for every possible pair ot attention level possibilities
        
        sub_split.loc[(sub_split.Subject == x) &(sub_split.CuedCategory == pair[0])&(sub_split.Category == pair[1]), 'Novs'] = float(novels[ (novels['Cued Category']==pair[0]) & (novels['Category']==pair[1]) ]['Familiarity Rating'])
        sub_split.loc[(sub_split.Subject == x)&(sub_split.CuedCategory == pair[1])&(sub_split.Category == pair[0]), 'Novs'] = float(novels[ (novels['Cued Category']==pair[1]) & (novels['Category']==pair[0]) ]['Familiarity Rating'])
        # add the appropriate novel image average rating
        
sub_split['Rating Differences'] = sub_split['Familiarity Rating']-sub_split['Novs']
```

### t-test
Differences between attention levels, all images considered together

```python
#l = sub_split['Attention Level'].unique()
l=['Category','None','Side','Full']

for pair in list(itertools.combinations(l, r=2)):
    t = stats.ttest_rel(sub_split[sub_split['Attention Level']==pair[0]]['Rating Differences'], 
                        sub_split[sub_split['Attention Level']==pair[1]]['Rating Differences'])
        
    # if t[1]>.05:
    print(pair[0]+" vs. "+pair[1])
    print(t) 
    print()
```

### t-test
Differences between attention levels, faces versus places

```python
print()
for attn in ['Category', 'Full', 'None', 'Novel', 'Side']:
    face = sub_split[(sub_split['Category']=='Place') & (sub_split['Attention Level']==attn)]['Rating Differences']
    place = sub_split[(sub_split['Category']=='Face') & (sub_split['Attention Level']==attn)]['Rating Differences']
    print(str(attn)+'  ->  '+str(stats.ttest_rel(face,place)))
    print(str('mean face '), str(face.mean()), str(face.std()))
    print(str('mean place '), str(place.mean()), str(place.std()))
    print()
    print()

figure = ax1.get_figure()    
figure.savefig('split_diffs.pdf', dpi=400)
```

# Gaze +/- 1.5

```python
def select_trials_2(behav_full, eye_df, width):
    '''
    input:   behavioral df, gaze df, desired width
    output:  trials with gaze within width of center (x axis)
    '''

    df_box = []
    gaze_three = []
    p_df = []

    for sub in eye_df['Subject'].unique():
    # for evey subject

        sub_df=[]

        chunk = eye_df[eye_df['Subject']==sub]
        # select data just for this subject

        for run in chunk['Run'].unique():
    #     # for every run

             for trial in chunk[chunk['Run']==run]['Trial'].unique():
    #         # for every trial

                 new_chunk = chunk[(chunk['Run']==run)&(chunk['Trial']==trial)]
    #             # select just the data for the run and trial

                 if new_chunk.shape[0] > 0 :
    #                 # if there's data for this run and trial

                    two = new_chunk[(abs(new_chunk['av_x_coord']-(59.8/2)) < width)].shape[0] / new_chunk.shape[0]
    #                 # what proportion of gaze points are within +/- 1.5 from center?

                    if two == 1:
                    # if 100%, these are trials we want

                        sub_df.append(pd.concat([behav_full[(behav_full['Subject']==sub) &
                                                 (behav_full['Run']==run) &
                                                  (behav_full['Trial']==trial)&
                                                  (behav_full['Trial Type']=='Presentation')]]))#,

    # #                                             behav_full[(behav_full['Subject']==sub) &
    # #                                             (behav_full['Run']==run) &
    # #                                             (behav_full['Trial Type']=='Memory')]]))

        sub_df.append(behav_full[(behav_full['Subject']==sub) & (behav_full['Trial Type']=='Memory')])

            # sub_df --> list of df's: [1] single good presentations, [2] all memory trials,


        if len(sub_df)>1:
            # if sub_df list has contents...

            sub_df = pd.concat(sub_df)
            df_box.append(sub_df)

            # concat into one df for the subject, then append to df_box
            # df box is a list of single sub df's containing 1) good pres trials and 2) possibly linked memory trials



    new_df_box = []

    for x in df_box:
        listy = list(x['Cued Place']) + list(x['Uncued Place']) + list(x['Cued Face']) + list(x['Uncued Face'])
        new_df_box.append(pd.concat([ x[x['Memory Image'].isin(listy)], x[x['Attention Level']=='Novel']]))
        ans = pd.concat(new_df_box)
        ans.ix[~ans['Memory Image'].str.contains("sun", na=False),'Category']='Face'
        ans.ix[ans['Memory Image'].str.contains("sun", na=False),'Category']='Place'

    return(ans)
```
