```python
def ROC_plot(ROC_data, save='a'):
    '''
    input: ROC proportions (dictionary)
    output: displays plot of ROC curve
    '''

    fig, ax = plt.subplots()

    for attn in ['Category', 'Full', 'None', 'Side']:
        ax.plot(ROC_data['Novel'], ROC_data[attn], '-o', label=attn)

    plt.legend(loc='upper left');
    plt.ylim(0, 1)
    plt.xlim(0, 1)
    plt.gca().set_aspect('equal', adjustable='box')
    
    fig.savefig(save)
    
    plt.show()
    
```

```python
# subject-wise ROC's, by category-cue block type

ROC_faceBlock_dict = {}
ROC_placeBlock_dict = {}

for ROC in ROC_df[0][0].columns[26:]:
    
    # for every ROC type
    ROC_faceBlock_dict[ROC],ROC_faceBlock_dict[ROC+'Cues'] = [],[]
    ROC_placeBlock_dict[ROC],ROC_placeBlock_dict[ROC+'Cues'] = [],[]
    
    for subject in ROC_df:
        # for every subject
        
        sub = pd.concat(subject)
        
        ROC_faceBlock_dict[ROC].append(pd.DataFrame(pd.DataFrame(
                                                    [sub[sub['Run']==x][ROC].iloc[0] for x in sub[sub['Cued Category'].str.contains("Face", na=False)]['Run'].unique()]).mean()).T)
        
        ROC_placeBlock_dict[ROC].append(pd.DataFrame(pd.DataFrame([sub[sub['Run']==x][ROC].iloc[0] 
                                     for x in sub[sub['Cued Category'].str.contains("Place", na=False)]['Run'].unique()]).mean()).T)
        

    ROC_faceBlock_dict[ROC] = pd.concat(ROC_faceBlock_dict[ROC])
    ROC_placeBlock_dict[ROC] = pd.concat(ROC_placeBlock_dict[ROC])
```
