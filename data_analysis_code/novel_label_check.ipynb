# Load Data

```python
exp1 = pd.DataFrame.from_csv('../parsed_data/behavioral_data_sustained.csv')
exp2 = pd.DataFrame.from_csv('../parsed_data/behavioral_data_variable.csv' )
```

# Data Organize

```python
# label rows by trial number
for exp in [exp1, exp2]:

    # Number all presentation and memory trials 
    exp.loc[exp['Trial Type']=='Memory','Trial'] = list(range(0,40))*30*8
    exp.loc[exp['Trial Type']=='Presentation','Trial'] = list(range(0,10))*30*8
```

### Add last-cued category from previous presentation block to memory trials

```python
for exp in [exp1,exp2]:
    
    exp['Last Cued'] = np.nan

    for sub in exp['Subject'].unique():
        for run in exp['Run'].unique():

            # obtain cued category from the last presentation trial
            last_cat = exp[(exp['Trial Type']=='Presentation')
                           & (exp['Subject']==sub)
                           & (exp['Run']==run)
                           & (exp['Trial']==9.0)]['Cued Category'].item()
            if last_cat != 'Face' and last_cat != 'Place':
                print(last_cat)

            # assign to this memory run
            exp.loc[(exp['Trial Type']=='Memory')
                           & (exp['Subject']==sub)
                           & (exp['Run']==run),'Last Cued'] = last_cat
```

### Re-Label Novel images by Cued or Uncued category

```python
exp1 = add_nov_label(exp1, column_name = 'Last Cued')
exp2 = add_nov_label(exp2, column_name = 'Last Cued')
```

## Check A1: Equal number of cued- and uncued-category images in each run for each participant

```python
for exp in [exp1, exp2]:
    check = []
    for s in exp['Subject'].unique():
        for r in exp['Run'].unique():
            data = exp[(exp['Subject']==s) & (exp['Run']==r)]
            check.append(data[data['Attention Level']=='Nov_Un'].shape == data[data['Attention Level']=='Nov_Cued'].shape)
    
    print('Is it true that in every run, for every subject, there are an equal number of cued- and uncued-cat novel images?')
    print('Answer: ' + str(set(check)))
    print()
```

## Check A2: In each run, only images from one category (Face or place) are labeled as uncued novel, and only images from the other category are labeled as cued novel

```python
for exp in [exp1, exp2]:
    for s in exp['Subject'].unique():
        for r in exp['Run'].unique():
            data = exp[(exp['Subject']==s) & (exp['Run']==r)]
            print(data[data['Attention Level']=='Nov_Un']['Category'].unique())
            print(data[data['Attention Level']=='Nov_Cued']['Category'].unique())
            
            if data[data['Attention Level']=='Nov_Un']['Category'].unique() == data[data['Attention Level']=='Nov_Cued']['Category'].unique():
                print('Novel images from the same category are labeled both as cued-cat and uncued-cat!')
                
            if len(data[data['Attention Level']=='Nov_Un']['Category'].unique())>1 or len(data[data['Attention Level']=='Nov_Cued']['Category'].unique())>1:
                print('Novel images from two categories both labeled as cued or both labeled as uncued!')
                
            print()
```

# Check A3: Randomly select two runs from two participants to spot check

```python
for exp_name,exp,seed in zip(['exp1','exp2'],[exp1,exp2],[3,4]):
    
    random.seed(seed)
    sub = random.sample(list(exp['Subject'].unique()), 2)
    run = random.sample(list(exp['Run'].unique()), 2)    

    for s,r in zip(sub,run):
        exp[(exp['Subject']==s) & (exp['Run']==r)].to_csv(exp_name+'_'+str(s)+'_'+str(r)+'_novelTest.csv')
```
