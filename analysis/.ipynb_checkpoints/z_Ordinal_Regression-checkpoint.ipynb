# Ordinal Regression

Here, we conduct an ordinal regression, treating the four familiarity ratings (1-4) as factors

# load packages

```r
library(lmerTest); library(lmerTest); library(lmerTest); library(lmerTest); library(ordinal); library(emmeans)
```

# load data

```r
dat <- read.csv("/Users/kirstenziman/Desktop/attention-memory-task/full.csv", header = TRUE)
dat <- within(dat, {Familiarity_Rating <- as.factor(Familiarity_Rating)})
dat <- dat[dat['Trial.Type']=='Memory',]
```

# ordinal regression

```r
mixed_model <- clmm(Familiarity_Rating ~ Gaze_Proportion + AL + (1|Subject), data = dat, threshold = "equidistant", Hess = TRUE)
summary(mixed_model)
```

```r
model.fixed = clm(Familiarity_Rating ~ Gaze_Proportion + AL, data = dat, threshold = "equidistant", Hess = TRUE)
anova(mixed_model,null = model.fixed)
```

# plot residuals

```r
plot(fitted(mixed_model))
```

# paired comparisons

```r
library(multcompView)
library(lsmeans)
library(ordinal)

marginal <- lsmeans(mixed_model, pairwise ~ AL, adjust="tukey") 

marginal
```
