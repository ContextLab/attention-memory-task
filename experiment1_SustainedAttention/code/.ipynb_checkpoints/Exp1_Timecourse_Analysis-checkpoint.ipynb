# Functions

```python
def apply_window(combo, window_length):
    '''
    input:  dataframe of behavioral data from an entire experiment
    output: dataframe of same shape where raw values have been replaced by rolling window mean  
            WARNING: THE FIRST WINDOW_LENGTH-1 CELLS IN OUTPUTTED DF ARE PADDED / NEED TO BE REMOVED BEFORE ANALYSIS?
    '''
    
    # select data from memory runs
    data = combo[combo['Trial Type']=='Memory'][['Attention Level','Familiarity Rating','Trial','Subject','Run']] 

    # re strucutre data - each row is a trial, each col is an attn level
    df = data.pivot_table(index=['Subject','Run','Trial'], 
                          columns='Attention Level', 
                         values='Familiarity Rating')

    # apply rolling window, for each run in each subject
    window_data = df.groupby(['Subject','Run']).apply(lambda x: x.rolling(window_length, min_periods=1).mean())

    return(window_data)
```

```python
def add_nov_label(combo, column_name = 'Cued Category'):
    '''
    input:  dataframe of participant data, and the name of the column to use for cue info
            for exp1 use 'Cued Category' (cue for that block)
            for exp2 use 'Last Cued'     (last cue from presentation block)
    output: dataframe where novel images are labeled by cued or uncued category
    '''

    for cat in ['Face','Place']:

        # rename all 'Novel' images --> 'Novel_Uncued'
        combo.loc[(combo['Trial Type']=='Memory') 
                  & (combo['Attention Level']=='Novel'),
                     'Attention Level'] = 'Nov_Un'

        # for all 'Novel_Uncued', if image in cued category, rename 'Novel_Cued'
        combo.loc[(combo['Trial Type']=='Memory') 
                  & (combo[column_name]==cat) 
                  & (combo['Category']==cat)
                  & (combo['Attention Level']=='Nov_Un'),
                     'Attention Level'] = 'Nov_Cued'
    return(combo)
```

# Sliding Window

```python
def sliding_window(data, wind_len, overlap):
    '''
    input : data (dataframe), window length, window overlap
    output: list of lists - one list for each window
    '''
    x = [data[i:i+wind_len] for i in range(0, len(data)-wind_len+1, overlap)]
    x = [a.index for a in x]
    
    return(x)
```

```python
def attn_index(data, run, attn_levels):
    '''
    input  : data, run, and category
    output : dictionary
            - keys : attention level
            - content : each key has a list of the indices that contain images from 
                        that attn level in mem run
    '''
    my_dict = {}
    
    for attn in attn_levels:
        my_dict[attn] = data[(data['Attention Level']==attn) & (data['Run']==run)].index
    
    
    return(my_dict)
```

```python
def run_familiarity(windows, indices, run_dat):
    '''
    input  : windows for this run, indices of attn levels, data for this run 
    output : dictionary 
             - keys : attention level
             - items: mean familiarity at each window point 
    '''
    
    familiarity = dict.fromkeys(indices.keys(),[])
    
    for w in windows:
    # for each window
        
        for attn in indices.keys():
        # for each attn level
            
            window_indices    = list(set(w) & set(indices[attn]))
            # find indices of images at this attn level in this window
            
            ratings = run_dat[(run_dat['Trial Type'] == 'Memory')]['Familiarity Rating']  
            # obtain ratings for this run
            
            familiarity[attn].append(ratings.loc[window_indices].mean()) 
            # append rating mean to appropriate dict key
            
    return(familiarity)
```

```python
def novel_index(data, subject, run, category, level='Novel'):
    '''
    input  : data, subject, run, and category
    output : indices of novel images from 'category' image category (F/P)
    '''
    selection = combo[(combo['Attention Level']==level) &(combo['Category']==category)
                            & (combo['Subject']==subject) & (combo['Run']==run)].index
    
    return(selection)
```

```python
def ratio_cued(windows, rating, face_nov_idx, place_nov_idx, cued_category, typ='mean'):
    '''
    input : for a given run: indices of sliding window bins, ratings, novel image indices, cued category
    output: Novel cued ratio for this run
    '''
    ratios = []
    
    for window in windows:
    # for each window

        face_win  = list(set(face_nov_idx) & set(window)) 
        place_win = list(set(place_nov_idx) & set(window))  
        # indices of novel face and place images in this window
        
        if typ =='mean':

            if cued_category == 'Face':
                ratio = rating[face_win].mean() #/ rating[place_win].mean()

            elif cued_category == 'Place':
                ratio = rating[place_win].mean() # / rating[face_win].mean()  #rating[face_win].mean()
                
        elif typ =='mean_ratio':

            if cued_category == 'Face':
                ratio = rating[face_win].mean() / rating[place_win].mean()

            elif cued_category == 'Place':
                ratio = rating[place_win].mean() / rating[face_win].mean()  #rating[face_win].mean()
            
            
        elif typ =='diff':
            ratio = abs(rating[place_win].mean() - rating[place_win].mean()) / (rating[face_win].mean() + rating[place_win])
          
        ratios.append(ratio)

    return(ratios)
```

```python
def ratio_uncued(windows, rating, face_nov_idx, place_nov_idx, cued_category, typ='mean'):
    '''
    input : for a given run: indices of sliding window bins, ratings, novel image indices, cued category
    output: Novel uncued ratio for this run
    '''
    ratios = []
    
    for window in windows:
    # for each window

        face_win  = list(set(face_nov_idx) & set(window)) 
        place_win = list(set(place_nov_idx) & set(window))  
        # indices of novel face and place images in this window
        
        if typ =='mean':

            if cued_category == 'Face':
                ratio = rating[place_win].mean() #/ rating[face_win].mean()

            elif cued_category == 'Place':
                ratio = rating[face_win].mean() #/ rating[place_win].mean()
            
        elif typ =='diff':
            ratio = abs(rating[face_win].mean() - rating[place_win].mean()) / (rating[face_win].mean() + rating[place_win])
          
        ratios.append(ratio)

    return(ratios)
```

```python
def ratio_all(windows, rating, attn_dict, typ='mean'):
    '''
    input : for a given run: indices of sliding window bins, ratings, novel image indices, cued category
    output: dict of ratios, each key is an attn level
    '''
    ratios = {}
    
    for window in windows:
    # for each window

        for k in attn_dict.keys(): 
        # for every attention level
            
            if (k != "Novel_Face") & (k != "Novel_Place") & (k!="Novel"):
        
                if typ =='mean':

                    ratio = rating[attn_dict].mean() #/ rating[face_win].mean()

                elif typ =='diff':
                    ratio = abs(rating[face_win].mean() - rating[place_win].mean()) / (rating[face_win].mean() + rating[place_win])

                ratios[k].append(ratio)

    return(ratios)
```

```python
def window(seq, n=2):
    "Returns a sliding window (of width n) over data from the iterable"
    "   s -> (s0,s1,...s[n-1]), (s1,s2,...,sn), ...                   "
    it = iter(seq)
    result = tuple(islice(it, n))
    if len(result) == n:
        yield result
    for elem in it:
        result = result[1:] + (elem,)
        yield result
```

# Gaze +/- 1.5

```python
def select_trials_2(behav_full, eye_df, width):
    '''
    input:   behavioral df, gaze df, desired width
    output:  trials with gaze within width of center (x axis)
    '''

    df_box = []
    gaze_three = []
    p_df = []

    for sub in eye_df['Subject'].unique():
    # for evey subject

        sub_df=[]

        chunk = eye_df[eye_df['Subject']==sub]
        # select data just for this subject

        for run in chunk['Run'].unique():
    #     # for every run

             for trial in chunk[chunk['Run']==run]['Trial'].unique():
    #         # for every trial

                 new_chunk = chunk[(chunk['Run']==run)&(chunk['Trial']==trial)]
    #             # select just the data for the run and trial

                 if new_chunk.shape[0] > 0 :
    #                 # if there's data for this run and trial

                    two = new_chunk[(abs(new_chunk['av_x_coord']-(59.8/2)) < width)].shape[0] / new_chunk.shape[0]
    #                 # what proportion of gaze points are within +/- 1.5 from center?

                    if two == 1:
                    # if 100%, these are trials we want

                        sub_df.append(pd.concat([behav_full[(behav_full['Subject']==sub) &
                                                 (behav_full['Run']==run) &
                                                  (behav_full['Trial']==trial)&
                                                  (behav_full['Trial Type']=='Presentation')]]))#,

    # #                                             behav_full[(behav_full['Subject']==sub) &
    # #                                             (behav_full['Run']==run) &
    # #                                             (behav_full['Trial Type']=='Memory')]]))

        sub_df.append(behav_full[(behav_full['Subject']==sub) & (behav_full['Trial Type']=='Memory')])

            # sub_df --> list of df's: [1] single good presentations, [2] all memory trials,


        if len(sub_df)>1:
            # if sub_df list has contents...

            sub_df = pd.concat(sub_df)
            df_box.append(sub_df)

            # concat into one df for the subject, then append to df_box
            # df box is a list of single sub df's containing 1) good pres trials and 2) possibly linked memory trials



    new_df_box = []

    for x in df_box:
        listy = list(x['Cued Place']) + list(x['Uncued Place']) + list(x['Cued Face']) + list(x['Uncued Face'])
        new_df_box.append(pd.concat([ x[x['Memory Image'].isin(listy)], x[x['Attention Level']=='Novel']]))
        ans = pd.concat(new_df_box)
        ans.ix[~ans['Memory Image'].str.contains("sun", na=False),'Category']='Face'
        ans.ix[ans['Memory Image'].str.contains("sun", na=False),'Category']='Place'

    return(ans)
```
