```python
sub_list=[]

for sub in df_list:
    lt = []
    for x in sub:
        lt.append(x.groupby(['Cued Face','Cued Place','Uncued Face','Uncued Place','Cued Side', 'Cued Category'], as_index=True)['av_x_coord'].mean())    
    df = pd.DataFrame(pd.concat(lt))
    
        
    df.columns = df.columns.get_level_values(0)
    df.columns = df.columns.map(''.join)
    df = df.reset_index()
    #sub_list2.append(df)
    df.ix[df['Cued Side']=='<','av_x_coord']=(df[df['Cued Side']=='<']['av_x_coord']*(-1))+(59.8/2.0)
    df.ix[df['Cued Side']=='>','av_x_coord']=df[df['Cued Side']=='>']['av_x_coord']-(59.8/2.0)

    sub_list.append(df)
```

```python
subs_rate=[]

for sub,mem in zip(sub_list2,behav_df):
 
    # make all distances in terms of distance TOWARD cued side (from center)
    sub[sub['Cued Side']=='<']['av_x_coord']=59.8/2.0-sub[sub['Cued Side']=='<']['av_x_coord']
    sub[sub['Cued Side']=='>']['av_x_coord']=sub[sub['Cued Side']=='>']['av_x_coord']-59.8/2.0
    
    sub['Cued Face Rating'] = np.nan
    sub['Cued Place Rating'] = np.nan
    sub['Uncued Face Rating'] = np.nan
    sub['Uncued Place Rating'] = np.nan
    
    for image,rate in zip(list(mem[mem['Trial Type']=='Memory']['Memory Image']),list(mem[mem['Trial Type']=='Memory']['Familiarity Rating'])):
        sub.ix[sub['Cued Face'].str.contains(image),'Cued Face Rating']=rate
        sub.ix[sub['Cued Place'].str.contains(image),'Cued Place Rating']=rate 
        sub.ix[sub['Uncued Place'].str.contains(image),'Uncued Place Rating']=rate 
        sub.ix[sub['Uncued Face'].str.contains(image),'Uncued Face Rating']=rate 
    #print(sub)
    
    subs_rate.append(sub)
```

```python
def pres_gaze(subdir, eye_df, interval='images'):
    '''
    input: subject's experiment df and eye track df
    output: list of eye data df's
            each df is either eye data from full pres block, or from single pres trial (interval='images')
    '''

    pres_gaze = []

    for f in os.listdir(subdir):

        stim=[]

        if 'pres' in f:

            pres_df = pd.read_csv(subdir+'/'+f)

            if interval == 'images':
                start = pres_df['Stimulus Onset']
                end = pres_df['Stimulus End']

            else:
                start = pres_df['Stimulus Onset'][0]
                end = pres_df['Stimulus End'][9]

            for x,y in zip(start,end):
                eye_df['Cued Side']=pres_df.iloc[0]['Cued Side']
                eye_df['Cued Category']=pres_df.iloc[0]['Cued Category']
                eye_df['Cued Face']=pres_df.iloc[0]['Cued Face']
                eye_df['Cued Scene']=pres_df.iloc[0]['Cued Scene']
                eye_df['Uncued Face']=pres_df.iloc[0]['Uncued Face']
                eye_df['Uncued Scene']=pres_df.iloc[0]['Uncued Scene']
                eye_df['Cued Category']=pres_df.iloc[0]['Cued Category']

                pres_gaze.append(eye_df.loc[(eye_df['timestamp']>x) &
                                            (eye_df['timestamp']<y) &
                                            (eye_df['xRaw_righteye']>0.0) &
                                            (eye_df['xRaw_lefteye']>0.0)])
                stim.append()
    return(pres_gaze)
```
